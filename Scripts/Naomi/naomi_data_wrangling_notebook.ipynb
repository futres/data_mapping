We can definitely work with this one. If there is a match between the two data sets then fill the bone type with the correct name 

```python
#Create bone_type column in deer_data
deer_data=deer_data.assign(bone_type_name="")

#Find match, if there's a match, assign bone type
for i in deer_data["bone_type_id"].index:
    for j in bone_type["bone_type_id"].index:
        if deer_data["bone_type_id"][i] == bone_type["bone_type_id"][j]:
            deer_data["bone_type_name"][i] = bone_type["bone_type_name"][j]
        
```

Do the same matching scheme again with the references

```python
#Create reference column in deer_data
deer_data=deer_data.assign(reference="")

#Find match, if there's a match, assign bone type
for i in deer_data["reference_id"].index:
    for j in reference["reference_id"].index:
        if deer_data["reference_id"][i] == reference["reference_id"][j]:
            deer_data["reference"][i] = reference["reference_details"][j]

#Note: a lot of the references are missing. 
```

Moving on to site.csv, which will have the same mechanisms, be a little more involved, but have a lot of pieces of the puzzle!

```python
site = pd.read_csv("site.csv")
```

```python
#Create site columns in deer_data
deer_data=deer_data.assign(site_name="")
deer_data=deer_data.assign(country_code="")
deer_data=deer_data.assign(region="")
deer_data=deer_data.assign(location="")

#Find match, if there's a match, assign bone type
for i in deer_data["site_code"].index:
    for j in site["site_code"].index:
        if deer_data["site_code"][i] == site["site_code"][j]:
            deer_data["site_name"][i] = site["site_name"][j]
            deer_data["country_code"][i] = site["country_code"][j]
            deer_data["region"][i] = site["region"][j]
            deer_data["location"][i] = site["location"][j]
```

Now we can move on to country code

```python
country = pd.read_csv("country.csv")
```

```python
#Create country column in deer_data
deer_data=deer_data.assign(country_name="")

#Find match, if there's a match, assign bone type
for i in deer_data["country_code"].index:
    for j in country["country_code"].index:
        if deer_data["country_code"][i] == country["country_code"][j]:
            deer_data["country_name"][i] = country["country_name"][j]
```

We can do the same method for species id 

```python
species = pd.read_csv("species.csv")
```

```python
#Create species column in deer_data
deer_data=deer_data.assign(common_name="")
deer_data=deer_data.assign(latin_name="")

#Find match, if there's a match, assign bone type
for i in deer_data["species_id"].index:
    for j in species["species_id"].index:
        if deer_data["species_id"][i] == species["species_id"][j]:
            deer_data["common_name"][i] = species["common_name"][j]
            deer_data["latin_name"][i] = species["latin_name"][j]
```

Moving on to period id 

```python
period = pd.read_csv("period.csv")
period
```

```python
#Create period columns in deer_data
deer_data=deer_data.assign(period_name="")
deer_data=deer_data.assign(sequence_number="")

#Find match, if there's a match, assign bone type
for i in deer_data["period_id"].index:
    for j in period["period_id"].index:
        if deer_data["period_id"][i] == period["period_id"][j]:
            deer_data["period_name"][i] = period["period_name"][j]
            deer_data["sequence_number"][i] = period["sequence_number"][j]
```
